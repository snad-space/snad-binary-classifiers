```python
import numpy as np
import time
import sys
sys.path.insert(0, '../..')

from rb_clf_V2.scripts.train_RBclf import load_single, get_akb_labels
```

```python
_, features = load_single('../../dr23-features/sid_snad_clf_r_100.dat', '../../dr23-features/feature_snad_clf_r_100.dat')
```

```python
t = time.monotonic()
#rb_proba = np.memmap('../../dr23-features/RBclf_proba.dat', mode='c', dtype=np.float32).reshape(-1,1)
sn_proba = np.memmap('../../dr23-features/SNclf_proba.dat', mode='c', dtype=np.float32).reshape(-1,1)
extended_features = np.hstack([features, sn_proba])

t = (time.monotonic() - t) / 60
print(f'Concat in {t:.0f}')
```

```python
# 1. Определяем имя файла и параметры
output_file = "../../dr23-features/features_sn.dat"
dtype = np.float32  # тот же dtype, что и у proba

# 2. Сохраняем массив в бинарном формате (memmap-совместимом)
t_start = time.monotonic()

# Создаем memmap-файл в режиме записи ('w+')
memmap_shape = extended_features.shape  # (N, M)
memmap_array = np.memmap(
    output_file,
    dtype=dtype,
    mode='w+',  # режим записи (создаст файл)
    shape=memmap_shape,
)

# Копируем данные из RAM в файл (построчно, если массив огромный)
memmap_array[:] = extended_features[:]

# Принудительно записываем изменения на диск
memmap_array.flush()
del memmap_array  # закрываем файл

t_total = (time.monotonic() - t_start) / 60
print(f"Массив сохранен в {output_file} за {t_total:.1f} мин.")
```
