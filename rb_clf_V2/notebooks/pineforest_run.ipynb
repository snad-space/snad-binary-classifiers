```python
from coniferest.pineforest import PineForest
from coniferest.session import Session
from coniferest.session.callback import TerminateAfter, viewer_decision_callback
import numpy as np
```

```python
def load_single(oid_filename, feature_filename):
    oid     = np.memmap(oid_filename, mode='c', dtype=np.uint64)
    feature = np.memmap(feature_filename, mode='c', dtype=np.float32).reshape(oid.shape[0], -1)
    return oid, feature
    
oids, features = load_single('dr23-features/sid_snad_clf_r_100.dat', 'dr23-features/feature_snad_clf_r_100.dat')
```

```python
feature_names = 'dr23-features/feature_snad_clf_r_100.name'
with open(feature_names) as f:
    names = f.read().split()

print(names)
```

```python
s = Session(features, oids,
           on_decision_callbacks= [TerminateAfter(5)],
           decision_callback=viewer_decision_callback,
           model=PineForest(random_seed=42))
```

```python
s.run()
```
