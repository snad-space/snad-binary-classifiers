```python
from coniferest.pineforest import PineForest
from coniferest.session import Session
from coniferest.session.callback import TerminateAfter, viewer_decision_callback
import numpy as np
import time
import pandas as pd

import sys
sys.path.insert(0, '../..')

from rb_clf_V2.scripts.utils import plot_config
from rb_clf_V2.scripts.train_RBclf import load_single, get_akb_labels
import time

plot_config()
```

```python
#oids, features = load_single('../../dr23-features/sid_snad_clf_r_100.dat', '../../dr23-features/feature_snad_clf_r_100.dat')
oids, features = load_single('../../dr23-features/collected_by_field_rb_no-chi2/dr23_oid_820.dat', '../../dr23-features/collected_by_field_rb_no-chi2/dr23_feature_820.dat')
```

```python
feature_names = '../../dr23-features/feature_snad_clf_r_100.name'
with open(feature_names) as f:
    names = f.read().split()

rs = 1
```

```python
s = Session(features[:,:-1], oids,
           on_decision_callbacks= [TerminateAfter(100)],
           decision_callback=viewer_decision_callback,
           model=PineForest(random_seed=rs))
```

```python
t = time.monotonic()
s.run()
t = (time.monotonic() - t) / 60
print(f'Session took {(t / 60):.0f}h. {t:.0f}m.')
```

```python
is_anomaly = np.asarray(list(s.known_labels.values()))
is_anomaly[is_anomaly == 1] = np.full(len(is_anomaly[is_anomaly == 1]), 0)
is_anomaly[is_anomaly == -1] = np.full(len(is_anomaly[is_anomaly == -1]), 1)
checked_oids = oids[list(s.known_labels.keys())]

table = pd.DataFrame(data={'oids': checked_oids, 'is_anomaly': is_anomaly})

table.to_csv(f'../pineforest_result/pineforest_initial_dr23_art_rs{rs}.csv', index=False)
```
