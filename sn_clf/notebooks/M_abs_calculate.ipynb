```python
import numpy as np
from astropy.cosmology import FlatLambdaCDM
from astropy import units as u
from astropy.coordinates import Distance


cosmo = FlatLambdaCDM(H0=70, Om0=0.3)

def compute_absolute_magnitude(m_r_peak, z, ebv):
    """
    Вычисляет абсолютную звездную величину в r-диапазоне.
    
    Параметры:
        m_r_peak (float): Видимая звездная величина
        z (float): Красное смещение
        ebv (float): Цветовой избыток E(B-V)
    
    Возвращает:
        float: Абсолютная звездная величина M_r_peak
    """
    A_r = 2.271 * ebv

    D = cosmo.luminosity_distance(z).to(u.pc).value
    #D = Distance(z=z, unit=u.pc).to_value()
    
    M_r_peak = m_r_peak - 5 * np.log10(D) + 5 #- A_r

    return M_r_peak

```

```python
z = 1.2
m_r_peak = 20
ebv = 0.03

M_r = compute_absolute_magnitude(m_r_peak, z, ebv)
print(f"Абсолютная звездная величина: M_r_peak = {M_r:.2f}")
```
