```python
from coniferest.isoforest import IsolationForest
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, '../..')

from sn_clf.scripts.utils import load_features, plot_config
```

```python
oids, features = load_features('../../dr23-features/sid_snad_clf_r_100.dat', '../../dr23-features/sn_rb.dat')
```

```python
plt.scatter(features[:10000, 1], features[:10000, 0], s=5, alpha=0.3)
```

```python
iforest = IsolationForest(random_seed=42).fit(features)
anom_score = iforest.score_samples(features)
```

```python
argsort = np.argsort(anom_score)
```

```python
oids[argsort[:100]]
```

```python
idx = argsort[:1000]
train_data = pd.read_csv('../data/train_data_big.csv')
mask = np.isin(oids[idx], train_data['oid'])
```

```python
cut = np.where(np.cumsum(~mask) == 101)[0][0]
```

```python
unique_cand_sn = oids[idx[:cut]][~np.isin(oids[idx[:cut]], train_data)]
corr_score = anom_score[idx[:cut]][~np.isin(oids[idx[:cut]], train_data)]
```

```python
urls = [f'https://ztf.snad.space/dr23/view/{oid}' for oid in unique_cand_sn]
ans = []
for i, url in enumerate(urls):
    print(url)
    a = int(input())
    ans.append(a)
```
