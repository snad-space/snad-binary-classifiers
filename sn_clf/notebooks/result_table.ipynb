```python
import os
import pandas as pd
import numpy as np
```

```python
out_name = [ name for name in os.listdir('../pineforest_output') if name[-3:]=='csv']
```

```python
fields = np.unique([name[-7:-4] for name in out_name])
```

```python
for field in fields:
    name = f'../pineforest_output/init_rs1_field{field}.csv'
    init = pd.read_csv(name).iloc[10:]

    name = f'../pineforest_output/aug_sn_rs1_field{field}.csv'
    aug = pd.read_csv(name).iloc[10:]

    top30 = pd.read_csv(f'../top30_snscore/field{field}.csv')
    
    init_count = init['is_anomaly'].sum()
    aug_count = aug['is_anomaly'].sum()
    top30_count = top30['is_sn'].sum()
    ind = np.argmax(np.array([init_count, aug_count, top30_count]))
    bf = [r'{', r'{', r'{']
    bf[ind] = r'\bf{'
    print(f"{field} & {bf[0]}{init_count:.0f}}} & {bf[1]}{aug_count:.0f}}} & {bf[2]}{top30_count:.0f}}} \\\\")
```
